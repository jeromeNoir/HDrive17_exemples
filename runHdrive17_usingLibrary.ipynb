# test amplitude at 0.5Hz

```python
frange=np.array([0.1,0.5,1,1.5,2])
amplitudeRange=np.array([10,50,100,200,300,400])

drivemode=134
maxAmplitude=np.zeros((frange.shape[0],amplitudeRange[0]))
minAmplitude=np.zeros((frange.shape[0],amplitudeRange[0]))
for _if,_f in enumerate(frange):
    for _ia,_amplitude in enumerate(amplitudeRange):
        duration=10*1/_f
        print('Frequency',_f)
        print('Amplitude',_amplitude)

        mydrive=hdrive()
        mydrive.connect_drive()
        mydrive.speed=2000
        mydrive.acc=30000
        mydrive.decc=30000
        mydrive.frequency=_f
        mydrive.mode=drivemode
        mydrive.amplitude=_amplitude
        mydrive.start() #start the rotation
        time.sleep(duration) 
        mydrive.stop()# stop the motor, home, reset
        mydrive.home()
        time.sleep(3)
        mydrive.servoOff()
        maxAmplitude[_if,_ia]=mydrive.act_pos.max()
        minAmplitude[_if,_ia]=mydrive.act_pos.min()
        mydrive.disconnect()
        time.sleep(3)
    
```
