### compare the motor position with the target position 

```python
figure, ax = subplots()
ax.plot(mydrive.act_time, mydrive.act_pos, label='Motor actual position')
ax.plot(mydrive.target_time, mydrive.target_pos, label='Target position')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Position (degrees)')
ax.set_title('Position vs Time')
ax.legend()
show()  
```
